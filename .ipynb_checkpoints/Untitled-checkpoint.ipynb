```python
# Imports for data manipulation
import numpy as np
import pandas as pd

# Imports for viewing
import matplotlib.pyplot as plt
import matplotlib as m
import seaborn as sns

# Imports for statistical calculations
import scipy
from scipy.stats import kurtosis, skew, shapiro
import warnings
warnings.filterwarnings("ignore")

# Imports for formatting graphics
plt.style.use('fivethirtyeight')
m.rcParams['axes.labelsize'] = 14
m.rcParams['xtick.labelsize'] = 12
m.rcParams['ytick.labelsize'] = 12
m.rcParams['text.color'] = 'k'
from matplotlib.pylab import rcParams 
rcParams['figure.figsize'] = 20,10
```

```python
df = pd.read_csv('SeoulBikeData.csv', encoding='latin1', parse_dates = ["Date"])
```

```python
# Replace space to underline
cols = df.columns
cols = cols.map(lambda x: x.replace(' ', '_') if isinstance(x, (str)) else x)
df.columns = cols

# checking dataset
df.head()
```

```python
df.info()
```

```python
plt.plot(df['Rented_Bike_Count'])

plt.title("Rent dayly ", size = 14)
plt.show()
```

```python
plt.bar(df['Seasons'], df['Rented_Bike_Count'])

plt.title("Rented_Bike_Count by Seasons ", size = 14)
plt.show()
```

```python
plt.bar(df['Hour'], df['Rented_Bike_Count'])

plt.title("Rented_Bike_Count by Hour ", size = 14)
plt.show()
```

```python
df.Hour.value_counts()
```

```python
# Vamos criar uma tabela pivot com Date, Seasons e Rented_Bike_Count
df_pivot = df.pivot_table(index = ['Date'], columns = ['Seasons'], values = 'Rented_Bike_Count')
```

```python
# Substituímos possíveis valores NA gerados no pivot, por 0 e transformamos o índice em coluna
df_pivot = df_pivot.fillna(0).reset_index()
```
